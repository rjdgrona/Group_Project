```python
%matplotlib inline
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
from scipy import stats

```

```python
#Read clean CSV
scatterdata_df = pd.read_csv("Output/cleaned_BMI_data.csv")
scatterdata_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)

```

```python
#Scatter Plot 2010-2013
sizes =[]
for x in scatterdata_df["Daily caloric supply (per person) (kcal/person/day)"]:
    sizes.append(x/200)
plt.scatter(x=scatterdata_df["% Adults > 25BMI"], y=scatterdata_df["Daily caloric supply (per person) (kcal/person/day)"], s=sizes, edgecolor="magenta", color="pink", alpha = .3)
# plt.grid()
plt.xlabel("Overweight & Obese Adults (%)")
plt.ylabel("Calories (kcal/person/day)")
plt.title("% Adults > 25BMI VS Daily caloric supply\n from 2010-2013")

#Linear Regression
x_axis = scatterdata_df["% Adults > 25BMI"]
y_axis = scatterdata_df["Daily caloric supply (per person) (kcal/person/day)"]

slope, intercept, r_value, p_value, std_err = stats.linregress(x_axis, y_axis)
print(stats.linregress(x_axis, y_axis))
print (f"r2 value {r_value*r_value}")

# plt.text(1,1,f"slope: {slope}\npvalue: {p_value}\nR-sq: {r_value*r_value}", bbox=dict(facecolor='red', alpha=0.5))
```

```python

#Specific Year
scatterdata_2013 = scatterdata_df.loc[scatterdata_df["Year"]== 2013, :]

#Linear Regression
x_axis_2013 = scatterdata_2013["% Adults > 25BMI"]
y_axis_2013 = scatterdata_2013["Daily caloric supply (per person) (kcal/person/day)"]

slope_2013, intercept_2013, r_value_2013, p_value_2013, std_err_2013 = stats.linregress(x_axis_2013, y_axis_2013)
r2_2013 = '{:.2f}'.format(r_value_2013*r_value_2013)
#Scatter Plot 
sizes =[]
for x in scatterdata_2013["Daily caloric supply (per person) (kcal/person/day)"]:
    sizes.append(x/100)
plt.figure(figsize=(6,4))    
plt.scatter(x=scatterdata_2013["% Adults > 25BMI"], y=scatterdata_2013["Daily caloric supply (per person) (kcal/person/day)"], s=sizes, edgecolor="magenta", color="pink", alpha = .5)
plt.grid(linestyle="--", alpha=.2)
plt.xlabel("Overweight & Obese Adults (%)")
plt.ylabel("Calories (kcal/person/day)")
plt.title("Percent Adult (BMI>25) & Daily Calorie Intake\n2013")

#USA
scatterdata_2013_USA = scatterdata_2013.loc[scatterdata_2013['Entity']=="United States",:]
plt.plot(scatterdata_2013_USA["% Adults > 25BMI"], scatterdata_2013_USA["Daily caloric supply (per person) (kcal/person/day)"], "o", alpha = 0.5, markerfacecolor='red', markeredgewidth=1.5, markeredgecolor="black")
plt.savefig("Output/BMI_calories.png")
plt.show()

#Print Linear Regression
print(f"slope: {'{:.2f}'.format(slope_2013)} \nintercept: {'{:.2f}'.format(intercept_2013)}\nr-sq: {r2_2013}\np value: {p_value_2013}")
```
