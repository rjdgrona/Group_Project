```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
adults_df = pd.read_csv("Resources/share-of-adults-who-are-overweight.csv")
adults_df = adults_df.dropna(how='any')
dates = [2010, 2011, 2012, 2013, 2014, 2015]
adults_df_2013 = adults_df[adults_df["Year"].isin(dates)]
# adults_df_2013 = adults_df.loc[adults_df["Year"] == 2013, :]
del adults_df_2013['Entity']
# del adults_df_2013['Year']
adults_df_2013 = adults_df_2013.rename(columns={"Indicator:Prevalence of overweight among adults, BMI &GreaterEqual; 25 (crude estimate) (%) - Age Group:18+  years - Sex:Both sexes (%)":"% Adults > 25BMI"})
adults_df_2013.head()
# adults_df_2013.columns
```

```python
cause_df = pd.read_csv("Resources/burden-of-disease-by-cause.csv")
cause_df = cause_df.dropna(how='any')
cause_df_2013 = cause_df[cause_df["Year"].isin(dates)]
del cause_df_2013['Entity']
# del cause_df_2013['Year']
cause_df_2013.head()
# cause_df_2013.columns
```

```python
men_bmi_file = "Resources/mean-body-mass-index-bmi-in-adult-males.csv"
men_bmi_df = pd.read_csv(men_bmi_file)
men_bmi_df = men_bmi_df.dropna(how='any')
men_bmi_df_2013 = men_bmi_df[men_bmi_df["Year"].isin(dates)]
del men_bmi_df_2013['Entity']
# del men_bmi_df_2013['Year']
del men_bmi_df_2013['Total population (Gapminder)']
men_bmi_df_2013.head()
```

```python
men_file = "Resources/share-of-adult-men-overweight-or-obese-vs-daily-supply-of-calories.csv"
men_df = pd.read_csv(men_file)
men_df = men_df.dropna(how='any')

men_df_2013 = men_df[men_df["Year"].isin(dates)]
men_df_2013.head()
```

```python
merge_men_table = pd.merge(men_df_2013, men_bmi_df_2013, on=["Code", "Year"] , how="left")
merge_men_table.columns
merge_men_table = merge_men_table.rename(columns={" (kg/m2)":"Men (BMI)"})
merge_men_table = merge_men_table.rename(columns={"Overweight or Obese (%)":"Men Overweight or Obese (%)"})
merge_men_table.head()
```

```python
women_bmi_file = "Resources/mean-body-mass-index-bmi-in-adult-women.csv"
women_bmi_df = pd.read_csv(women_bmi_file)
women_bmi_df = women_bmi_df.dropna(how='any')
women_bmi_df_2013 = women_bmi_df[women_bmi_df["Year"].isin(dates)]

del women_bmi_df_2013['Entity']
# del women_bmi_df_2013['Year']
del women_bmi_df_2013['Total population (Gapminder)']
women_bmi_df_2013.head()
```

```python
women_file = "Resources/share-of-adult-women-overweight-or-obese-vs-daily-caloric-supply.csv"
women_df = pd.read_csv(women_file)
women_df = women_df.dropna(how='any')
women_df_2013 = women_df[women_df["Year"].isin(dates)]
del women_df_2013['Entity']
# del women_df_2013['Year']
del women_df_2013['Daily caloric supply (per person) (kcal/person/day)']
del women_df_2013['Total population (Gapminder)']
women_df_2013.head()
```

```python
merge_women_table = pd.merge(women_bmi_df_2013, women_df_2013, on=["Code", "Year"], how="left")
merge_women_table.columns
merge_women_table = merge_women_table.rename(columns={" (kg/m2)":"Women (BMI)"})
merge_women_table = merge_women_table.rename(columns={"Share of women overweight or obese (%)":"Women overweight or obese (%)"})
merge_women_table.head()
```

```python
merge_table = pd.merge(merge_men_table, merge_women_table, on=["Code", "Year"], how="left")
merge_table = merge_table[["Entity","Code","Year",
                           "Men Overweight or Obese (%)",
                           "Men (BMI)",
                           "Women overweight or obese (%)",
                           "Women (BMI)",
                           "Daily caloric supply (per person) (kcal/person/day)",
                           "Total population (Gapminder)"]]
merge_table
```

```python
merge_table = pd.merge(merge_table, cause_df_2013, on=["Code", "Year"], how="left")
merge_table.head()
```

```python
merge_table = pd.merge(merge_table, adults_df_2013, on=["Code", "Year"], how="left")
merge_table.head()
```

```python
merge_table.to_csv("Output/cleaned_BMI_data.csv", index=False)
```
